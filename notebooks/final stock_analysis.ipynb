# Data Import and Preparation

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

#Reading the time series data of the index an 'United Bank Of India EOD Prices' and different stocks of Indian Stock Exchange
# The data we are getting are seperate for each stock, hence we are reading them seperately and then
nifty50 = pd.read_csv("^NSEI.csv")

#Auto sector
eicher = pd.read_csv("EICHERMOT.NS.csv")
tata = pd.read_csv("TATAMOTORS.NS.csv")
mahindra = pd.read_csv("M&M.NS.csv")
maruti = pd.read_csv("MARUTI.NS.csv")

# Pharma sector
piramal = pd.read_csv("PEL.NS.csv")
sun = pd.read_csv("SUNPHARMA.NS.csv")
reddy = pd.read_csv("RDY.csv")
apollo = pd.read_csv("APOLLOHOSP.NS.csv")

#Banking secotr
sbi = pd.read_csv("SBIN.NS.csv")
icici = pd.read_csv("ICICIBANK.NS.csv")
hdfc = pd.read_csv("HDB.csv")
bajaj = pd.read_csv("BAJFINANCE.NS.csv")
```

```python
stocks_list = [nifty50, eicher, tata, mahindra, maruti, piramal, sun, reddy, apollo, sbi, icici, hdfc, bajaj]
```

```python
stocks_list
```

```python
stocks_list_names = ['nifty50', 'eicher', 'tata', 'mahindra', 'maruti', 'piramal', 'sun', 'reddy', 'apollo', 'sbi', 'icici', 'hdfc', 'bajaj']
```

```python
# Manually creating a dictionary object type to have values of SectorName column which we will create in further steps

sector_dict = {'nifty50': 'index',
               'eicher': 'auto',
               'tata': 'auto', 
               'mahindra': 'auto',
               'maruti': 'auto',
               'piramal': 'health',
               'sun': 'health',
               'reddy': 'health',
               'apollo': 'health',
               'sbi': 'banking',
               'icici': 'banking',
               'hdfc' : 'banking',
               'bajaj': 'banking'}
```

```python
# Printing out data information of each stock dataframe 

for stock in stocks_list:
    print(stock.info())
```

```python
# Since we got different dataframes for each stock it becomes imperative to create a stock name column to identify each stock
# Creating a for loop to create a StockName and SectorName column which will contain value of the name of dataframe and their corresponding sector as in dictionary we created

for index, dataframe in enumerate(stocks_list):
    dataframe['StockName'] = stocks_list_names[index]
    dataframe['SectorName'] = sector_dict[stocks_list_names[index]]
    print(dataframe)
```

```python
#  concatenating all the dataframes of different stocks into into single dataframe

df = pd.concat(stocks_list, axis=0)

# axis=0 specifies that the concatenation should be done vertically (row-wise)
# If the dataframes have the same column names, they will be aligned automatically
# If the dataframes have different column names, NaN values will be filled for missing columns

# Reset the index of the concatenated dataframe
df = df.reset_index(drop=True)
df
```

```python
# Dropping the column Adj Close as we don't see much relevannce of that in our analysis
# We are creating another variable called Intra Day Variation which measures 
# how much difference in a day was between closing and opening price of a share

df.drop(columns=['Adj Close'], inplace=True)
df['IntraDay_Variation'] = df['Close'] - df['Open']
df
```

```python
df.describe().T
```

```python
df[df.SectorName == 'health'].describe().T
```

```python
# WE want descriptive statistics but not of the full data as the data is based on different categorical features
# Like Stock Name and Stock Industry
# So we calculate and compare descriptive statistics based on Sector Name and StockNames

for sector in set(df.SectorName):
    print(f'Description of Sector: {sector}')
    print(df[df.SectorName == f'{sector}'].describe().T)
```

```python
for stockname in set(df.StockName):
    print(f'Description of Sector: {stockname}')
    print(df[df.StockName == f'{stockname}'].describe().T)
```

```python
df
```

```python
# Creating a column called Base Price which is Close price of a stock at first day 
# Setting the first day's value of each stock name as the base price

for stockname in set(df.StockName):
    sub_df = df[df.StockName == f'{stockname}']
    sub_df.reset_index(drop=True, inplace=True)
    first_value = sub_df.Close[0]
    df.loc[df.StockName == f'{stockname}', 'BasePrice'] = first_value
```

```python
# Just making sure that the column contains same amount of unique values as number of stock names

df.BasePrice.nunique()
```

```python
# Now we have Base Price column populated with initial price of each stock name
# we calcuate Percentage Change of each day's Close Price of each stock

df['Percentage_Change'] = (df.Close - df.BasePrice) / (df.BasePrice) * 100
df
```
