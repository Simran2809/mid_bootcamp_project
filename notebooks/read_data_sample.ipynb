```python
import pandas as pd

print(pd.__version__)
```

```python
#Reading the time series data of the index 'Nifty 50' for the financial year starting on 01.04.2013 upto 31.03.2023
nifty50 = pd.read_csv("../data/raw/^NSEI.csv")
nifty50.head(15)
```

```python
# Determining the data types of each column and evaluating whether they are appropriate for the nature of the variable.

print(nifty50.columns)
print(type(nifty50.columns))
print(nifty50.dtypes)
```

```python
# Identify the number of unique values for each column.

unique_values_nifty50 = nifty50.nunique()
print(unique_values_nifty50)
```
